```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the tickers we are interested in
tickers = ['UNH', 'NOC', 'ORCL', 'GE', 'MPC']

# Loop through each ticker
for ticker in tickers:
    # Download the historical stock price data
    stock_data = yf.download(ticker, start='2020-01-01', end='2023-01-01')

    # Calculate the 50-day simple moving average (SMA50)
    stock_data['SMA50'] = stock_data['Close'].rolling(window=50).mean()

    # Calculate the 200-day simple moving average (SMA200)
    stock_data['SMA200'] = stock_data['Close'].rolling(window=200).mean()

    # Start plotting
    plt.figure(figsize=(12,8))

    # Plot the closing price
    plt.plot(stock_data['Close'], label='Close Price')

    # Plot the SMA50
    plt.plot(stock_data['SMA50'], label='SMA50')

    # Plot the SMA200
    plt.plot(stock_data['SMA200'], label='SMA200')

    # Add a title to the plot
    plt.title(f'{ticker} Stock Price and Moving Averages')

    # Add a legend to the plot
    plt.legend()

    # Show the plot
    plt.show()
```
