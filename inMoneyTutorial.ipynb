```python
# Create a Tickers object for multiple tickers
tickers = yf.Tickers('msft aapl goog')

# Create a dictionary to store the info of each ticker
info_dict = {}


for ticker in tickers.tickers:
    info_dict[ticker] = tickers.tickers[ticker].info

# Convert the dictionary to a DataFrame
info_df = pd.DataFrame.from_dict(info_dict, orient='index')

# Display the DataFrame
info_df
```

```python
#optional
msft_info = tickers.tickers['MSFT'].info
df_msft = pd.DataFrame.from_dict(msft_info, orient='index', columns=['MSFT'])
df_msft
```
