```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the tickers we are interested in
tickers = ['UNH', 'NOC', 'WMT', 'QCOM', 'CWCO', 'KR']

# Loop through each ticker
for ticker in tickers:
    # Download the historical stock price data
    stock_data = yf.download(ticker, start='2010-01-01', end='2023-01-01')

    # Calculate the 50-day simple moving average (SMA50)
    stock_data['SMA50'] = stock_data['Close'].rolling(window=50).mean()

    # Calculate the 200-day simple moving average (SMA200)
    stock_data['SMA200'] = stock_data['Close'].rolling(window=200).mean()

    # Start plotting
    plt.figure(figsize=(12,8))

    # Plot the closing price
    plt.plot(stock_data['Close'], label='Close Price')

    # Plot the SMA50
    plt.plot(stock_data['SMA50'], label='SMA50')

    # Plot the SMA200
    plt.plot(stock_data['SMA200'], label='SMA200')

    # Add a title to the plot
    plt.title(f'{ticker} Stock Price and Moving Averages')

    # Add a legend to the plot
    plt.legend()

    # Show the plot
    plt.show()
```

```python
KR = yf.download('KR')
kr = KR[['Close']].copy()  # Ensure it's a DataFrame with a single column

# Calculate moving averages
kr['SMA50'] = kr['Close'].rolling(window=50, min_periods=50).mean()
kr['SMA200'] = kr['Close'].rolling(window=200, min_periods=200).mean()

# When using `ewm`, the `min_periods` should typically be set equal to `span` or omitted
# to allow the calculation from the beginning of the series
kr['EMA100'] = kr['Close'].ewm(span=100, adjust=False).mean()
kr
```

```python
kr.plot(figsize=(12,8), fontsize=15)
plt.legend(loc='upper left', fontsize=15)
plt.show()
```

# Cummulative Returns and Drawdowns

```python
apple=yf.download('AAPL')
```

```python
apple=apple.Close.to_frame()
apple
```

```python
apple.Close
```

```python
apple['d_returns']=np.log(apple.div(apple.shift(1)))
```

```python
apple
```

```python
apple.dropna(inplace=True)
```

```python
apple
```

```python
apple.d_returns.sum()
```

```python
np.exp(apple.d_returns.sum())
```

```python
apple['culmreturns']=apple.d_returns.cumsum().apply(np.exp)
```

```python
apple
```

```python
apple.culmreturns.plot(figsize=(12,8), title='AAPLE buy and hold', fontsize=12)
plt.show()
```

```python
apple.d_returns.mean()*252
```

```python
apple.d_returns.std()*np.sqrt(252)
```

```python
# how to calculate drawdowns
apple['culmmax']=apple.culmreturns.cummax()
```

```python
apple
```

```python
apple[['culmreturns','culmmax']].plot(figsize=(12,8), title='AAPLE buy and hold + Culmmax', fontsize=12)
plt.show()
```

```python
apple['drawdown']=apple['culmmax']-apple['culmreturns']
```

```python
apple
```

```python
apple.drawdown.max()
```

```python
apple.drawdown.idxmax()
```

## SMA Strategy

```python
data= apple.Close.loc[(apple.index>='1991-01-01')]
```

```python
data=data.to_frame()
```

```python
data
```

```python
smaSmall=50
smaLarge=100
```

```python
data['smaSmall']=data.Close.rolling(smaSmall).mean()
data['smaLarge']=data.Close.rolling(smaLarge).mean()
```

```python
data
```

```python
data.plot(figsize=(12,8),title='AAPLE - SMA{} | SMA{}'.format(smaSmall,smaLarge),fontsize=12)
```

```python
data.loc['2016'].plot(figsize=(12,8),title='AAPLE - SMA{} | SMA{}'.format(smaSmall,smaLarge),fontsize=12)
```

```python
data.dropna(inplace=True)
data
```

```python
data['position']=np.where(data['smaSmall']>data['smaLarge'],1,0)
```

```python
data.loc['2016',['smaSmall', 'smaLarge', 'position']].plot(figsize=(12,8),title='AAPLE - SMA{} | SMA{}'.format(smaSmall,smaLarge),fontsize=12,secondary_y='position')
```

```python
data['returnsB&H']=np.log(data.Close.div(data.Close.shift(1)))
```

```python
data['strategy']=data['returnsB&H']*data.position.shift(1)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
#Comparison Buy and Hold
```

```python
data[['returnsB&H', 'strategy']].sum()
```

```python
data[['returnsB&H', 'strategy']].sum().apply(np.exp) #what 1$ will be
```

```python
data[['returnsB&H', 'strategy']].std()*np.sqrt(252) #annual std
```

```python
data.Close.plot(figsize=(12,8))
```
