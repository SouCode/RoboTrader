```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
```

```python
topPicks = ['SPY','UNH', 'NOC', 'NVDA', 'MSFT', 'PANW', 'V', 'TSLA',
            'BA', 'WMT', 'MPC', 'ABNB', 'GE', 'QCOM', 'ORCL', 'DIS', 
           'OKTA', 'BHP', 'SLB', 'TSN', 'EQT']
stocks = yf.download(topPicks, start = '2008-01-01', end ='2023-10-01')
```

```python
stocks
```

```python
close = stocks.loc[:,'Close'].copy()
topClose = close.div(close.iloc[0]).mul(100)
```

```python
topClose.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show
```

```python
stats = close.pct_change().dropna()
stats.head()
```

```python
stats.describe().T
```

```python
summary = stats.describe().T.loc[:,['mean', 'std']]
summary
```

```python
summary['mean'] = summary['mean']*252
summary['std'] = summary['std']*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x='std',y='mean',figsize=(12,8),s=50,fontsize=15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,'std']+0.002,summary.loc[i,'mean']+0.002),size=15)
plt.xlabel('Annual risk(std)', fontsize=15)
plt.ylabel('Annual return', fontsize=15)
plt.title('Risk/return', fontsize=25)
plt.show()
```

```python
stats.corr()
```

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=1.4)
sns.heatmap(stats.corr(), cmap='Reds',annot=True, annot_kws={'size':12}, vmax=0.6)
plt.show()
```

```python
SPY = yf.download('SPY')
spy = SPY[['Close']].copy()  # Ensure it's a DataFrame with a single column
spy.rename(columns={'Close': 'SPY'}, inplace=True)
spy
```

```python
spy.plot(figsize=(12,8), fontsize=15)
plt.legend(loc='upper left', fontsize=15)
plt.show()
```

```python
spy_roll=spy.rolling(window=10)
spy_roll=spy_roll.mean().dropna()
spy_roll.head(15)
```

```python
spy_roll=spy.rolling(window=10).median().dropna()
spy_roll
```

```python
spy_roll=spy.rolling(window=10).max().dropna()
spy_roll
```

```python
spy_roll=spy.rolling(window=10,min_periods=5).max().dropna()
spy_roll
```

```python
spy_roll=spy.rolling(window=10,min_periods=5).max().head(15).dropna()
spy_roll
```

```python
spy
```

```python
spy['SMA50']=spy.rolling(window=50,min_periods=50).mean()
#adding dropna breaks it
spy
```

```python
spy.plot(figsize=(12,8), fontsize=15)
plt.legend(loc='upper left', fontsize=15)
plt.show()
```
